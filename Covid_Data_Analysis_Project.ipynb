```python
covid.columns
```

```python
covid.drop(['Sno','Time','ConfirmedIndianNational','ConfirmedForeignNational'],inplace = True,axis = 1)
```

```python
covid.head()
```

## Active cases

```python
covid['Active_Cases'] = covid['Confirmed'] - (covid['Cured'] + covid['Deaths'])
```

```python
covid.tail()
```

```python
state_wise = pd.pivot_table(covid,values = ['Confirmed','Deaths','Cured'],index = 'State/UnionTerritory',aggfunc = max)
```

## Recovery rate

```python
state_wise['Recovery Rate'] = state_wise['Cured']*100/state_wise['Confirmed']
```

## Mortality rate

```python
state_wise['Mortality Rate'] = state_wise['Deaths']*100/state_wise['Confirmed']
```

```python
state_wise = state_wise.sort_values(by = 'Confirmed',ascending = False)
```

## Plotting pivot table

```python
state_wise.style.background_gradient(cmap = 'cubehelix')
```

## Top 10 active cases states

```python
top_10_active_cases = covid.groupby(by = 'State/UnionTerritory').max()[['Active_Cases','Date']].sort_values(by = ['Active_Cases'],ascending = False).reset_index()
fig = plt.figure(figsize=(16,9))

plt.title("Top 10 states with most active cases in India",size = 25)
ax = sns.barplot(data = top_10_active_cases.iloc[:10],y = 'Active_Cases',x = 'State/UnionTerritory',linewidth = 2,edgecolor = 'black')

plt.xlabel("States")
plt.ylabel("Total Active Cases")
```

## Top states with highest deaths

```python
top_10_deaths = covid.groupby(by = 'State/UnionTerritory').max()[['Deaths','Date']].sort_values(by = ['Deaths'],ascending = False).reset_index()
fig = plt.figure(figsize=(18,5))

plt.title("Top 10 states with most deaths",size = 25)
ax = sns.barplot(data = top_10_deaths.iloc[:12],y = 'Deaths',x = 'State/UnionTerritory',linewidth = 2,edgecolor = 'black')

plt.xlabel("States")
plt.ylabel("Total Death Cases")
```

## Remove rows where state = India

```python
vacc = vaccine[vaccine.State!='India']
```

```python
vacc.head()
```

```python
vacc.rename(columns = {'Total Individuals Vaccinated':'Total'},inplace = True)
```

```python
vacc.head()
```

## Most Vaccinated States

```python
max_vac = vacc.groupby('State')['Total'].sum().to_frame('Total')
max_vac = max_vac.sort_values('Total',ascending = False)[:5]
max_vac
```

```python
fig = plt.figure(figsize=(10,5))

plt.title("Top 5 Vaccinated States in India",size = 20)
x = sns.barplot(data = max_vac.iloc[:10],y = max_vac.Total,x = max_vac.index,linewidth = 2,edgecolor = "black")

plt.xlabel("States")
plt.ylabel("Vaccination")
```
